```python
import cv2
import matplotlib.pyplot as plt
```

## Resize the image

```python
# Orignal image
img = cv2.imread("Images/pic_1.jpg")
plt.imshow(img)
plt.axis("off")
plt.show()
print(img.shape)
```

```python
# Resized image
# interpolation = cv2.INTER_AREA --> This interpolation is usefull while shrinking the image
# origin img size was (173,291) here shriking to (150,150)
resize_img = cv2.resize(img,(150,150), interpolation = cv2.INTER_AREA)
plt.imshow(resize_img)
plt.axis("off")
plt.show()
print(resize_img.shape)
```

* Some case if we want enlarge the img from smaller size to bigger size then 
* interpolation has to be change 
* interpolation of enlarge --> 1. cv2.INTER_LINEAR 2. cv2.INTER_CUBIC 
* INTER_CUBIC interpolation gives much higher quality than INTER_LINEAR interpolation 

```python
img.shape
```

```python
# origin img size was (173,291) here enlarging to (250,250)

#linear interpolation
resize_img_inter_linear = cv2.resize(img,(250,250), interpolation = cv2.INTER_LINEAR)
resize_img_inter_cubic = cv2.resize(img,(250,250), interpolation = cv2.INTER_CUBIC)

# plt.subplot(2,1,1)
plt.imshow(resize_img_inter_linear)
plt.title("Enlarge img with linear interpolation")
plt.xticks([])
plt.yticks([])
plt.axis("on")
plt.xlabel(resize_img.shape)
plt.show()
# plt.subplot(2,2,1)
plt.imshow(resize_img_inter_cubic)
plt.title("Enlarge img with cubic interpolation")
plt.xticks([])
plt.yticks([])
plt.axis("on")
plt.xlabel(resize_img.shape)

plt.show()
```

### Croping

* Basically images are arrays of picxles
* We implifies the array slicing method to crop the images means we can select the portion of image on the bases of pickle values

```python
from PIL import Image
```

```python
image = cv2.imread("Images/car.jpg") # image converted into arrays
Image.open("Images/car.jpg")
```

```python
## Displaying a Matplotlib RGB image but results are in BGR because Open cv reads images in BGR 
## When we want to display images in matplotlib library we convert it into BGR TO RGB
plt.imshow(image) #
plt.show()
```

## Displaying a Matplotlib RGB Image
### Note
* OpenCV represents RGB images as multi-dimensional NumPy arrays…but in reverse order!
* This means that images are actually represented in BGR order rather than RGB!
* All we need to do is convert the image from BGR to RGB:
* 

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) #
plt.show()
```

## Want to crop the image of a car 
* Find the car pickel values 
* Using image slicing method to crop 
* car pickles values x1 = 90, x2=260 , y1 = 60, y2= 130

```python
cropped = image[100:260,60:130]
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)) #
plt.show()
```
