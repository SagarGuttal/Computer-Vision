# Color spaces in Computer Vision
* There are several different color spaces and each has its own significance. Some of the popular color spaces are RGB (Red, Green, Blue), CMYK (Cyan, Magenta, Yellow, Black), HSV (Hue, Saturation, Value), etc. BGR color space: OpenCV's default color space is RGB. However, it actually stores color in the BGR format.20-Apr-2022
* Computer vision has default color space that is --> BGR

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("Images/peple.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

### Convert BGR to gray-scale image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap="gray")
plt.axis("off")
plt.show()
```

### Convert BGR to HSV (hue, saturation, value)

```python
# bgr_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
HSV = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(HSV)
plt.axis("off")
plt.show()
```

### Convert BGR TO LAB

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2.imshow("lab",lab)
cv2.waitKey(0)
```

## Color channels

```python
img =  cv2.imread("Images/person.webp")

#Spliting color channels
b,g,r = cv2.split(img)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB))
plt.title("Blue split image")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(cv2.cvtColor(g, cv2.COLOR_BGR2RGB))
plt.title("green split image")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(cv2.cvtColor(r, cv2.COLOR_BGR2RGB))
plt.title("red split image")
plt.axis("off")
plt.show()

print("Shape of original image :-", img.shape)
print("Shape of blue split image :-", b.shape)
print("Shape of green split image :-", g.shape)
print("Shape of red split image :-", r.shape)
```

## Merge the color channels

```python
merged = cv2.merge([r,g,b])
plt.imshow(merged)
plt.title("red split image")
plt.axis("off")
plt.show()
print("Shape of merged images :-",merged.shape)
```

## Visualizing separate color channels

```python
img = cv2.imread("Images/colorfull images.jpg")
blank = np.zeros(img.shape[:2], dtype="uint8")
b,g,r = cv2.split(img)

blue_img = cv2.merge([b,blank,blank])
green_img = cv2.merge([blank,g,blank])
red_img = cv2.merge([blank,blank,r])

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("original image")
plt.axis("off")
print("Shape of original image :-",img.shape)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(blue_img, cv2.COLOR_BGR2RGB))
plt.title("blue image")
plt.axis("off")
print("Shape of blue image :-",blue_img.shape)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(green_img, cv2.COLOR_BGR2RGB))
plt.title("green image")
plt.axis("off")
print("Shape of green image :-",green_img.shape)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB))
plt.title("red image")
plt.axis("off")
print("Shape of red image :-",red_img.shape)
```

```python
# merging all color channels
merged = cv2.merge([red_img,green_img,blue_img])
plt.figure(figsize=(6,6))
plt.imshow(merged)
plt.title("all color channels merged image")
plt.axis("off")
print("Shape of blue image :-",merged.shape)
```
