# Image transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Images/Boy pic.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
def translate(img, x, y):
    transmat = np.float32([[1,0,x],[0,1,y]]) #Create translation image
    dimensions = (img.shape[1],img.shape[0])
    
    return cv2.warpAffine(img, transmat, dimensions)
```

* -x(negative of x value) --> Left
* -y(negative of y value) --> up
* +x (positive of x value)---> Right
* +Y(positive of y vaue) ---> Down

```python
# Shift the image 10 pixcel by right and 20 pixcel by down
translated_img = translate(img, 10, 20)
plt.imshow(cv2.cvtColor(translated_img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Shift the image 10 pixcel by left and 20 pixcel by up
translated_img = translate(img, -10, -20)
plt.imshow(cv2.cvtColor(translated_img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## Image rotation

```python
def rotate(img, angle, rotpoint=None):
    (height,width) = img.shape[:2]
    
    if rotpoint is None:
        rotpoint = (width//2,height//2) # Center of rotstion point
        
    rotMat = cv2.getRotationMatrix2D(rotpoint, angle, 1.0) #Create rotation metrix
    dimensions = (width,height)
    
    return cv2.warpAffine(img, rotMat, dimensions)
```

```python
# Clockwise rotation when it is positive values
rotated = rotate(img, 45)
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Anti-Clockwise rotation when it is negative values
rotated = rotate(img, -45)
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## Fliping the image

* flip_code = 0 for fliping the image vertically over the x-axis
* flip_code = 1 for fliping the image horizontally over the y-axis
* flip_code = -1 for fliping the image both vertically and horizontally 

```python
flip_image = cv2.flip(img,0)
plt.imshow(cv2.cvtColor(flip_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
flip_image = cv2.flip(img,1)
plt.imshow(cv2.cvtColor(flip_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
flip_image = cv2.flip(img,-1)
plt.imshow(cv2.cvtColor(flip_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
